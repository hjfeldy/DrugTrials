```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

#Define rounding function
def roundTwo(statistic):
    return round(statistic, 2)

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
df = pd.merge(study_results, mouse_metadata, on='Mouse ID')

# Display the data table for preview
df
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

mean = df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
med = df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
var = df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
std = df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem = df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

summ = pd.DataFrame({'Mean': mean, 'Median': med, 'Variance': var, 'std': std, 'SEM': sem})
summ.applymap(roundTwo) #prettify with the custom round function
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

agg = df.groupby('Drug Regimen').agg(['mean', 'median', 'var', 'std', 'sem'])['Tumor Volume (mm3)']
agg.applymap(roundTwo) #prettify
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin



# Start by getting the last (greatest) timepoint for each mouse
last = df.groupby('Mouse ID').max()['Timepoint']


# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
df = df.merge(last, on='Mouse ID').rename(columns = {'Timepoint_y': 'Last Measurement', 'Timepoint_x': 'Timepoint'})


treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
#Sort list to preserve order between groupbys and reference df
treatments.sort()

#Filter for only the Last Measurement and the 4 desired drugs
final = df[(df['Last Measurement'] == df['Timepoint']) & (df['Drug Regimen'].isin(treatments))]

final = final[['Mouse ID', 'Tumor Volume (mm3)', 'Drug Regimen']]
final
```

```python
#The instructions in this cell were so poorly worded, I have no idea how you want me to find the outliers.
#So I'm just gonna do it a sensible way

#1. perform a groupby on the "final" df (final tumor volumes), grouping by the selected drugs
#Function to aggregate is the quantile function(obviously)
volumes = final.groupby('Drug Regimen').quantile([.25, .75])

volumes
```

```python
#2. Translate the ugly multi-index frame into a reference dataframe 
    #we will perform a merge later to identify outliers this way

#I'm gonna make two columns (lists), one for each quantile
bot = []
top = []

for row in volumes.iterrows():
    tile = row[0][1]
    val = row[1][0]
    if tile == .25:
        bot.append(val)
    else:
        top.append(val)
                                                #Note the "index=treatments"
#Crate reference df and IQR column              if we didn't do this, the above for loop would be out-of-whack
ref = pd.DataFrame({'25th': bot, '75th': top}, index=treatments) #This is why we sorted the treatments list before
ref['IQR'] = ref['75th'] - ref['25th']                            #Because groupbys always sort alphabetically
ref                                                          
```

What a nice, useful little dataframe! 
Since we made sure to set the index to the ALPHABETIZED list of treatments, we know that our values (which we obtained from a multi-index groupby) will line up correctly

```python
#We can now quantify outliers with the 1.5*IQR rule
ref['Big Outlier'] = 1.5*ref['IQR'] + ref['75th']
ref['Small Outlier'] = ref['25th'] - 1.5 * ref['IQR']

#Now that we have these columns, we don't need the IQR or the percentiles anymore. 
#We only needed those to obtain the outlier thresholds
ref = ref[['Big Outlier', 'Small Outlier']]
ref
```

```python
#Now we can perform a merge with our reference df and identify all outliers quantitatively

#Seriously, the instructions that were commented into the notebook were awful. 
#It would have been better to just stop at "Find the IQR and identify outliers quantitatively" and give us freedom

final = pd.merge(final, ref, left_on='Drug Regimen', right_index=True)
final['Outlier?'] = ( (final['Tumor Volume (mm3)'] > final['Big Outlier']) | (final['Tumor Volume (mm3)'] < final['Small Outlier']) )

final
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

# matplotlib's boxplot function requires that you pass a LIST OF dataframes/series 
# in order to get everything on one plot 
# You can't just call plt.boxplot multiple times, like you can with plt.plot


# Here is out list of series
ser = []

for treatment in treatments:
   ser.append(final[final['Drug Regimen'] == treatment]['Tumor Volume (mm3)'])

plt.boxplot(ser, labels= treatments)
plt.show()
```

```python
#Check if the outlier I found quantitatively is the same as the one that matplotlib found
final[final['Outlier?'] == True]
```

Note there is only one outlier, and the drug is Infubinol

Everything lines up. Nice!

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

#picked m601 at random from the df[df['Drug Regimen'] == 'Capomulin'] slice
tumorTime = df[df['Mouse ID'] == 'm601'] 
tumorTime.plot('Timepoint', 'Tumor Volume (mm3)')
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
weight_and_volume = df[df['Drug Regimen'] == 'Capomulin'].groupby('Mouse ID').mean()
weight_and_volume.plot(kind='scatter', x='Weight (g)', y='Tumor Volume (mm3)')
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
x = weight_and_volume['Weight (g)']
y = weight_and_volume['Tumor Volume (mm3)']

res = st.linregress(x, y=y)


rsq = res[2]
print(f'Correlation Coefficient = {roundTwo(rsq)}')
```

## Visualization of Regression Fit

```python
intercept = res[1]
slope = res[0]
weight_and_volume['Prediction'] = intercept + slope * weight_and_volume['Weight (g)']

weight_and_volume.plot(kind='scatter', x='Weight (g)', y='Tumor Volume (mm3)')
plt.plot(weight_and_volume['Weight (g)'] , weight_and_volume['Prediction'])
plt.show()
```
